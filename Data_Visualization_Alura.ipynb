```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/refs/heads/main/Dados/imigrantes_canada.csv'
df= pd.read_csv(url , sep=',')

df
```

```python
df.info()
```

```python
df.set_index('País' , inplace=True)
```

```python
years = list(map(str , range(1980,2014)))
```

```python
years
```

```python
data_br = df.loc['Brasil' , years]
data_br.head(4)
```

```python
data_ag = df.loc['Argentina' , years]
data_ag.head(5)
```

```python


agentina_dict = {"year": data_ag.index.to_list(), 'immigrants': data_ag.values.tolist()}

data_arg = pd.DataFrame(agentina_dict)
data_arg.head(5)
```

```python
brasil_dict = {"year": data_br.index.to_list(), 'immigrants': data_br.values.tolist()}

data_brasil = pd.DataFrame(brasil_dict)
data_brasil.head(5)
```

#Gráfico 1

```python
plt.figure(figsize=(8,4))
plt.plot(data_brasil['year'] , data_brasil['immigrants'])
plt.xticks(['1980' , '1985','1990','1995', '2000', '2005', '2010'])

plt.xlabel("Years")
plt.ylabel("Number of Immigrants")
plt.title("Number of Brazilian immigrants to Canada (1980–2013)")



plt.show()
```

```python
data = df.loc[['Argentina' ,'Brasil'], years]
```

```python
data
```

```python
# transpose
data = data.T
```

```python
data.head(5)
```

#Gráfico 2

```python
plt.title("Number of Brazilian and Argentine immigrants to Canada (1980–2013)")

plt.plot(data['Brasil'], label = 'Brasil', color ='y')
plt.plot(data['Argentina'] , label = 'Argentina')

plt.xticks(['1980' , '1985','1990','1995', '2000', '2005', '2010'])
plt.xlabel("Years")
plt.ylabel("Number of immigrants")

plt.legend()


plt.show()
```

#Gráfico 3

```python
# fig > figura , ax > eixos
fig, ax = plt.subplots(figsize=(8,4))

ax.plot(data_brasil['year'],data_brasil['immigrants'])
ax.set_title("Brazilian immigrants to Canada 1980–2013" , fontsize=12 , loc='left')
ax.set_xlabel("Years" , fontsize=12)
ax.set_ylabel("Number of immigrants")
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

ax.xaxis.set_major_locator(plt.MultipleLocator(5))

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

fig.savefig('imigração_brasil_canada.png',transparent=False, dpi=300 , bbox_inches='tight')
plt.show()
```

#Gráfico 5

```python
fig , axs = plt.subplots(2,2 , figsize=(15,10))
# espaçamento entre figs hspace=0.5 horizontal ,wspace=0.3 vertical
fig.subplots_adjust(hspace=0.4,wspace=0.7)
fig.suptitle("Immigration from the 4 largest countries \n in South America to Canada 1980 - 2013", fontsize=15)

axs[0,0].plot(df.loc['Brasil' , years] , color='g')
axs[0,0].set_title("Brasil")

axs[0,1].plot(df.loc["Colômbia", years] , color='y')
axs[0,1].set_title("Colômbia")

axs[1,0].plot(df.loc["Argentina",years])
axs[1,0].set_title("Argentina")

axs[1,1].plot(df.loc["Peru",years] , color='red')
axs[1,1].set_title("Peru")

# loop para peccorer os eixos
for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_ylabel("Years")
  ax.set_ylabel("Number of immigrants")
  ax.spines['top'].set_visible(False)
  ax.spines['right'].set_visible(False)

ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)

plt.show()
```

```python
america_sul = df.query('Região == "América do Sul"')
america_sul = america_sul.sort_values('Total', ascending=True)

america_sul.head(2)
```

#Gráfico 6

```python
cores = []
# for para add a cor verde apenas no brasil e cinza no resto
for pais in america_sul.index:
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('silver')

fig , ax = plt.subplots(figsize=(12,5))

ax.barh(america_sul.index, america_sul['Total'] , color=cores)
ax.set_title("América do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canadá no período de 1980 a 2013", loc='left', fontsize=16)
ax.set_xlabel("Número de imigrantes", fontsize=14)
ax.set_ylabel("")
ax.yaxis.set_tick_params(labelsize=12) # tamanho da label do eixo y e x
ax.xaxis.set_tick_params(labelsize=12)

for index , imigrantes in enumerate(america_sul['Total']):
  # função text para add textos na figura
  # 20 pixels
  # str para converter o numero de imigrantes em string
  ax.text(imigrantes + 20, index , str(imigrantes) , color='k' , fontsize=10, ha='left' , va='center')

ax.set_frame_on(False) #remover o frame(linha)

ax.get_xaxis().set_visible(False) #visibilidade do eixo x
ax.tick_params(axis='both', which='both' , length=0) # removendo o eixo y(traços) , both > ambos


plt.show()
```
